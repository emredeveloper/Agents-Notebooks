```python
from pathlib import Path

telegram_path = Path("/content/nanobot/nanobot/channels/telegram.py")
code = telegram_path.read_text()

old = '''                else:
                    content_parts.append(f"[{media_type}: {file_path}]")

                logger.debug(f"Downloaded {media_type} to {file_path}")'''

new = '''                elif ext == ".pdf" or getattr(media_file, "mime_type", "") == "application/pdf":
                    try:
                        import pypdf as _pypdf
                        reader = _pypdf.PdfReader(str(file_path))
                        pdf_text = "\\n".join(p.extract_text() or "" for p in reader.pages)
                        if pdf_text.strip():
                            content_parts.append(f"[PDF içeriği ({len(reader.pages)} sayfa):\\n{pdf_text[:8000]}]")
                        else:
                            content_parts.append(f"[PDF: metin çıkarılamadı - {file_path}]")
                    except Exception as _e:
                        content_parts.append(f"[PDF okuma hatası: {_e}]")
                else:
                    content_parts.append(f"[{media_type}: {file_path}]")

                logger.debug(f"Downloaded {media_type} to {file_path}")'''

if old in code:
    telegram_path.write_text(code.replace(old, new))
    print("✅ Patch başarılı!")
else:
    print("❌ Bulunamadı, şu satırları göster:")
    for i, line in enumerate(code.split('\n')):
        if 'content_parts.append' in line and ('media_type' in line or 'file_path' in line):
            print(f"  {i}: {repr(line)}")
```

Local ollama with ngrok tunnel endpoint

```python
import json, os

cfg_path = os.path.expanduser("~/.nanobot/config.json")

with open(cfg_path) as f:
    cfg = json.load(f)

cfg.setdefault("providers", {})
cfg["providers"].setdefault("custom", {})

# ngrok ollama endpoint
cfg["providers"]["custom"]["apiBase"] = \
    "ngrok url"

# dummy key (ollama kullanmıyor ama config istiyor)
cfg["providers"]["custom"]["apiKey"] = "ollama-local"

# model
cfg["agents"]["defaults"]["model"] = "granite4:3b"

with open(cfg_path, "w") as f:
    json.dump(cfg, f, indent=2)

print("✅ Ollama provider fixed")
```
